```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Nutrical Dataset.csv')

# Display basic information
print(df.info())
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill or drop missing values if necessary
df = df.dropna()  # Example of dropping missing values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Calories'], bins=20, kde=True)
plt.title('Calorie Count Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(df[['Calories', 'Total Fat', 'Protein', 'Carbohydrates']])
plt.show()
```

```python
# Bar chart of average calories per category
category_calories = df.groupby('Category')['Calories'].mean().sort_values()
category_calories.plot(kind='barh', color='skyblue')
plt.title('Average Calories per Category')
plt.xlabel('Average Calories')
plt.show()

# Box plot of calories in different categories
sns.boxplot(x='Category', y='Calories', data=df)
plt.xticks(rotation=90)
plt.title('Calorie Distribution by Category')
plt.show()
```

```python
highest_calorie = df.loc[df['Calories'].idxmax()]
lowest_calorie = df.loc[df['Calories'].idxmin()]
print('Highest Calorie Item:', highest_calorie)
print('Lowest Calorie Item:', lowest_calorie)
```

```python
# Bar chart comparing nutritional content across categories
plt.figure(figsize=(12, 8))
sns.barplot(x='Category', y='Calories', data=df, estimator=lambda x: sum(x) / len(x))
plt.xticks(rotation=45)
plt.title('Average Calories by Category')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Summary statistics of numerical columns
print(df.describe())

# Analyze the distribution of calorie counts
plt.figure(figsize=(10, 6))
sns.histplot(df['Calories'], bins=30, kde=True)
plt.title('Distribution of Calorie Counts')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()

# Explore nutritional content (e.g., fat, protein, carbohydrates)
nutritional_columns = ['Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']
for column in nutritional_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], bins=30, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Boxplot of calorie distribution by food category
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Calories', data=df)
plt.title('Calorie Distribution by Food Category')
plt.xticks(rotation=45)
plt.show()

# Compare nutritional characteristics of different food categories
for column in nutritional_columns:
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Category', y=column, data=df)
    plt.title(f'Average {column} by Food Category')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Identify menu items with the highest and lowest calorie counts
highest_calories = df.loc[df['Calories'].idxmax()]
lowest_calories = df.loc[df['Calories'].idxmin()]
print("Menu Item with Highest Calories:\n", highest_calories)
print("Menu Item with Lowest Calories:\n", lowest_calories)

# Determine the average nutritional content of popular menu categories
avg_nutrition_by_category = df.groupby('Category')[nutritional_columns].mean()
print("Average Nutritional Content by Category:\n", avg_nutrition_by_category)
```
